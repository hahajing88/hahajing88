```python
import pandas as pd 
import json 
import os
import copy
import charset_normalizer
```

```python

# 创建一个空的列表，用于存储所有JSON文件的数据
merged_data = []

# 遍历每个JSON文件
for filename in os.listdir("./data/图情学科"):
    if filename.endswith('.json'):
        # 读取JSON文件
        with open(os.path.join("./data/图情学科", filename), 'r', encoding='utf-8') as file:
            for line in file:
                try:
                    data = json.loads(line)
                    # 将数据添加到合并的数据列表中
                    merged_data.append(data)
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON in file {filename}: {e}")

# 打印合并后的数据
print(merged_data)
```

```python
# 将合并后的数据保存为新的JSON文件
with open('./data/图情学科/all_data_图情学科.json', 'w', encoding='utf-8') as file:
    json.dump(merged_data, file, ensure_ascii=False, indent=4)
```

```python
# 针对合并后的json文件的数据类型转换

# 读取JSON文件
with open('./data/图情学科/all_data_图情学科.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 创建空DataFrame存储实体和类型
df = pd.DataFrame(columns=['编号','索引', '实体类型', '实体'])

# 遍历每个JSON文件的数据
for file_index, file_data in enumerate(data):
    # 遍历每个索引和对应的实体类型和实体值
    for index, entities in file_data.items():
        # 计算编号
        if file_index == 0:
            number = int(index) + 1
        else:
            number = int(index) + (file_index * 20) + 1
        # print(file_index)
        #遍历每个实体类型和实体值
        for entity_type, entity_values in entities.items():
            # 添加到DataFrame里
            for entity_value in entity_values:
                df = pd.concat([df, pd.DataFrame({'编号': [number],'索引':[index], '实体类型': [entity_type], '实体': [entity_value]})])

df.reset_index(drop=True, inplace=True)

# 保存数据
df.to_excel('./data/图情学科/entities_all_图情学科.xlsx', index=False)
```
