```python
import pandas as pd

# 读取数据
df = pd.read_excel('./data/图情学科/副本entities_all_图情学科(加编号)1040处理后.xlsx',sheet_name='图情学科-处理后1040')

# 创建一个空列表，用于存储生成的三元组
triples = []
```

```python
# 创建空的 DataFrame 用于存储三元组
triples_df = pd.DataFrame(columns=['实体1', '实体1类型', '关系', '实体2', '实体2类型'])

# 遍历每个索引
for index in df['序号'].unique():
    # 根据索引筛选相应的实体
    entities = df[df['序号'] == index]

    # 确定实体类型
    entity_types = entities['实体类型'].unique()

    # 根据实体类型进行三元组映射
    for entity_type in entity_types:
        entity_values = entities[entities['实体类型'] == entity_type]['实体'].tolist()

        if entity_type == '研究问题':
            for value in entity_values:
                # 判断是否有实体类型为方法模型的实体
                if '方法模型' in entity_types:
                    method_models = entities[entities['实体类型'] == '方法模型']['实体'].tolist()
                    for method_model in method_models:
                        triples_df = pd.concat([triples_df, pd.DataFrame({'实体1': [value], '实体1类型': [entity_type], '关系': ['采用'], '实体2': [method_model], '实体2类型': ['方法模型']})], ignore_index=True)

                # 判断是否有实体类型为理论原理的实体
                if '理论原理' in entity_types:
                    theory_principles = entities[entities['实体类型'] == '理论原理']['实体'].tolist()
                    for theory_principle in theory_principles:
                        triples_df = pd.concat([triples_df, pd.DataFrame({'实体1': [value], '实体1类型': [entity_type], '关系': ['基于'], '实体2': [theory_principle], '实体2类型': ['理论原理']})], ignore_index=True)

        if entity_type == '数据资料':
            for value in entity_values:
                # 判断是否有实体类型为研究问题的实体
                if '研究问题' in entity_types:
                    research_questions = entities[entities['实体类型'] == '研究问题']['实体'].tolist()
                    for research_question in research_questions:
                        triples_df = pd.concat([triples_df, pd.DataFrame({'实体1': [value], '实体1类型': [entity_type], '关系': ['辅助解决'], '实体2': [research_question], '实体2类型': ['研究问题']})], ignore_index=True)

        if entity_type == '方法模型':
            for value in entity_values:
                # 判断是否有实体类型为软件系统的实体
                if '软件系统' in entity_types:
                    software_systems = entities[entities['实体类型'] == '软件系统']['实体'].tolist()
                    for software_system in software_systems:
                        triples_df = pd.concat([triples_df, pd.DataFrame({'实体1': [value], '实体1类型': [entity_type], '关系': ['借助'], '实体2': [software_system], '实体2类型': ['软件系统']})], ignore_index=True)

        if entity_type == '度量指标':
            for value in entity_values:
                # 判断是否有实体类型为研究问题的实体
                if '研究问题' in entity_types:
                    research_questions = entities[entities['实体类型'] == '研究问题']['实体'].tolist()
                    for research_question in research_questions:
                        triples_df = pd.concat([triples_df, pd.DataFrame({'实体1': [value], '实体1类型': [entity_type], '关系': ['测度'], '实体2': [research_question], '实体2类型': ['研究问题']})], ignore_index=True)

# 保存三元组数据至 Excel 文件
triples_df.to_excel('./data/图情学科/triples_图情学科1040.xlsx', index=False)
```
