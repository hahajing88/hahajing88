```python
import jieba
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence

# 加载自定义词典
jieba.load_userdict("专业词表_keywords_all.txt")

# 加载停用词表
stopwords = []
with open("哈工大停用词表.txt", "r", encoding="utf-8") as f:
    stopwords = [line.strip() for line in f]
```

```python
# 分词和去除停用词函数
def seg_and_remove_stopwords(text):
    seg_list = jieba.cut(str(text))
    seg_list = [seg for seg in seg_list if seg not in stopwords and len(seg.strip()) > 0]
    return seg_list
```

```python
import pandas as pd
data=pd.read_excel('./data/数据_预处理后.xlsx')
data.head()
```

```python
corpus=data['Summary-摘要'].tolist()
# 处理摘要数据
corpus = [seg_and_remove_stopwords(text) for text in corpus]
```

```python
corpus
```

```python
# 训练Word2Vec模型
model = Word2Vec(sentences=corpus, vector_size=300, window=5, min_count=1, sg=0, epochs=20)

# 保存模型
model.save("word2vec_model.bin")
```

```python
# 研究问题的相似度计算
# 加载预训练的词向量模型
model = Word2Vec.load("word2vec_model.bin")
# 读取Excel文件
df = pd.read_excel("./data/数据_图情学科1040、科学学和传播学.xlsx")

df.head()
```

```python
# 筛选实体类型为"研究问题"的实体
research_questions = df.loc[df["实体类型"] == "研究问题", "实体"].tolist()
research_questions
```

```python
# 分词处理主题领域研究问题并获取词向量
# research_question = "中西部地区基层公共图书馆人员保障机制"
research_question ="智能化时代的科学学"
s1_list = seg_and_remove_stopwords(research_question)
research_question_vector = [model.wv[word] for word in s1_list if word in model.wv]
research_question_vector
```

```python
# 计算研究问题与其他学科研究问题的相似度
similarity_scores = []
# subjects = []
for question in research_questions:
    # 分词处理其他学科的研究问题并获取词向量
    tokens=seg_and_remove_stopwords(question)
    question_vector = [model.wv[word] for word in tokens if word in model.wv]
    # print(question_vector)
    # 计算相似度
    try:
        similarity = model.wv.n_similarity(research_question_vector, question_vector)
        if similarity is None:
            similarity = 0  # 将相似度为空时赋值为0
    except ZeroDivisionError:
        similarity = 0 
    similarity_scores.append(similarity)
    # print(similarity)
```

```python
# 将相似度添加到DataFrame中
df_filtered = df.loc[df["实体类型"] == "研究问题"].copy()
df_filtered["相似度"] = similarity_scores

# 提取研究问题对应的学科信息
# df_subjects = df_filtered[["实体", "学科"]].copy()

# 打印DataFrame
print(df_filtered)
# 保存数据
df_filtered.to_excel('./data/图情学科/实例2_相似度计算.xlsx', index=False)
```
